0 \le t \le 50
$$

```python
t = np.linspace(0, 50, 1000)
x = t*np.cos(t)
y = t*np.sin(t)
plt.plot(x, y)
plt.axis('equal')
pass
```

 
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1
$$

```python
a = 2
b = 1
x = np.linspace(-a, a, 300)
y = np.linspace(-b, b, 300)
X, Y = np.meshgrid(x, y)
Z = X**2/a**2 + Y**2/b**2
plt.contour(X, Y, Z, 1)
plt.set_cmap('viridis')
plt.axis('equal')
pass
```

```python
x = np.linspace(-1, 1, 300)
X, Y = np.meshgrid(x, x)
Z = (X**2 + Y**2)**3 - 4*X**2*Y**2
cv = [-0.1, -0.05, -0.01, -0.001, -0.00005, 0, 0.1]
plt.contour(X, Y, Z, cv)
plt.set_cmap('viridis')
plt.colorbar()
pass
```

```python
plt.contourf(X, Y, Z, cv)
pass
```

```python
X, Y = np.meshgrid(np.arange(-3, 3, .1), np.arange(-3, 3, .1))
Z = np.sin(X) * np.sin(Y)
cs = plt.contour(X, Y, Z)
plt.clabel(cs, fontsize=10)
plt.set_cmap('hsv')
plt.colorbar()
pass
```

   
$(r, \varphi)$
$$
r = \sin(a\varphi), \qquad 0\le\varphi\le 2\pi
$$

```python
plt.figure(figsize = (10, 8))
phi = np.linspace(0, 2*np.pi, 200)
plt.polar(phi, np.sin(4*phi))
plt.polar(phi, np.cos(2*phi))
pass
```

```python
plt.figure(figsize = (10, 8))
phi = np.linspace(0, 2*np.pi, 200)
plt.polar(phi, np.abs(np.sin(4*phi)))
plt.polar(phi, np.abs(np.cos(2*phi)))
pass
```
