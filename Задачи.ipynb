## 38. Выплаты по кредиту. Решение трансцендентного уравнения

Предположим, что кто-то желает взять кредит на сумму $\$100000$.
Он выбирает среди следующих возможностей:

 1) взять кредит на $10$ лет с ежемесячными выплатами в $\$1500$;
 
 2) взять кредит на $15$ лет с ежемесячными выплатами в $\$1250$;
 
 3) взять кредит на $20$ лет с ежемесячными выплатами в $\$1160$.

Какой кредит выгоднее брать?

Можно просто вычислить переплату, а можно еще учесть процентные ставки.

Сами процентные ставки здесь не указаны, но можно их найти из *основного финансового уравнения*:

$$
{\rm PV}\, (1+i)^n + {\rm PMT}\, (1+i)^{n-1} + {\rm PMT}\, (1+i)^{n-2} + \dots + {\rm PMT} + {\rm FV} = 0, \tag{1}
$$
связывающего следующие величины:
 - ${\rm PV}$ - величина кредита (present value);
 - ${\rm FV}$ - баланс после $n$ платежей (future value);
 - ${\rm PMT}$ - величина периодического платежа со знаком минус (payment);
 - $n$ - количество платежей;
 - $i$ - процентная ставка (interest rate).
 
Например, для случая 1):
 - ${\rm PV} = 100000$;
 - ${\rm FV} = 0$ (после всех платежей банк и частное лицо друг другу ничего не должны);
 - ${\rm PMT} = -1500$;
 - $n = 10\times 12 = 120$.

# 38. Функция Ламберта

*Функция Ламберта* (или product log - функция) $W(x)$ - это функция, обратная к функции
$$
y = x e^x. \tag{1}
$$

Вначале построим график функции $x e^x$:

````python
xx = np.linspace(-7, 1)
yy = xx*np.exp(xx)
plt.plot(xx, yy)
plt.title('$x e^x$')
plt.grid()
````

График обратной функции теперь проще построить, поменяв местами аргументы в функции `plt.plot()`:

````python
plt.plot(yy, xx)
plt.title('$W(x)$')
plt.grid()
````

Мы видим, что функция Ламберта - многозначная. В области вещественных чисел у нее две ветви (и счетное число ветвей, если рассматривать комплекснозначные функции комплексного аргумента): 

 - "верхняя", определенная для всех $x \ge -1/e$, при этом $y\ge -1$, и 
 - "нижняя" - для $-1/e\le x < 0$, при этом $y\le -1$.
 
Изобразим своим цветом каждую из этих ветвей:

````python
xxm1 = np.linspace(-7, -1)
yym1 = xxm1*np.exp(xxm1)

xx0 = np.linspace(-1, 1)
yy0 = xx0*np.exp(xx0)

plt.plot(yy0, xx0, label='$k=0$')
plt.plot(yym1, xxm1, label='$k=-1$')
plt.title('$W(x)$')
plt.legend()
plt.grid()
````

Функцию Ламберта можно вычислить с помощью функции `scipy.special.lambertw`:

 - `lambertw(x, k=0)` дает значения для "верхней" ветви (по умолчанию);
 - `lambertw(x, k=-1)` дает значения для "нижней" ветви.
 

````python
from scipy.special import lambertw

lambertw(0, k=0), lambertw(1), lambertw(2.5), lambertw(-.2, k=-1).real
````

Вычисления этой функции могут быть не очень точны вблизи точки $x_0 = -1/e$. В частности, график этой функции можно нарисовать с помощью `lambertw(x)`, но они получаются не такими аккуратными, как  были построены выше.

````python
lambertw(-np.exp(-1)), lambertw(-np.exp(-1)+1e-6), lambertw(-np.exp(-1), k=-1), lambertw(-np.exp(-1)+1e-6, k=-1)
````
