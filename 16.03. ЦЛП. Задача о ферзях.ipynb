# Задача о ферзях

```python
import pulp
```

```python
prob = pulp.LpProblem("The Queens Problem", pulp.LpMinimize)
```

```python
#prob += 0, "Arbitrary Objective Function" 
```

```python
n = 8
rows = range(n)
cols = range(n)
positions = pulp.LpVariable.dicts("Position", (rows, cols), 0, 1, pulp.LpInteger)
```

```python
positions[0][3]
```

## Ограничения

В каждом столбце и каждой строке ровно по одному ферзю:

```python
for r in rows:
    prob += pulp.lpSum([positions[r][c] for c in cols]) == 1
for c in cols:
    prob += pulp.lpSum([positions[r][c] for r in rows]) == 1    
```

На каждой диагонали не более одного ферзя:

```python
for d in range(1, n):
    prob += pulp.lpSum([positions[r][d - r] for r in range(d + 1)]) <= 1
for d in range(n - 2):
    prob += pulp.lpSum([positions[r][d - r + n] for r in range(d + 1, n)]) <= 1    
```

```python
for d in range(n - 1):
    prob += pulp.lpSum([positions[r][d + r] for r in range(n - d)]) <= 1
for d in range(1, n - 1):
    prob += pulp.lpSum([positions[d + r][r] for r in range(n - d)]) <= 1    
```

## Решаем задачу

```python
#prob.writeLP("QueensModel.lp")
```

```python
prob.solve()
#print("Status:", pulp.LpStatus[prob.status])
```

```python
def printSolution():
    print('-----------------')
    for r in rows:
        for c in cols:
            if pulp.value(positions[r][c]) == 0:
                print(' .', end='')
            else:
                print(' Q', end='')
        print()   
    print('-----------------')        
```

```python
printSolution()
```

# Как найти все решения?

```python
nSolutions = 0
while True:
    if prob.solve() == 1:
        nSolutions += 1
        printSolution()
        prob += pulp.lpSum([positions[r][c] for r in rows 
                                            for c in cols
                                            if pulp.value(positions[r][c]) == 1]) <= n - 1
    else:
        break
```

```python
print("Number of solutions:", nSolutions)
```
